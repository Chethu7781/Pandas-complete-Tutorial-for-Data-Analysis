## Making changes to the data

```python
# Creating a new column of total and adding values
df['total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
# manual way
```

```python
# drop a specific columns
df = df.drop(columns=['total'])
df.head(5)
```

```python
# Creating a column and adding using iloc method
df['total'] = df.iloc[:,4:10].sum(axis=1) # [all rows = :, columns from 4 to 10]
```

```python
# Importing regex it is an superfast tool for data analysis 
import re

#df.loc[df['Type 1'].str.contains('Grass|Rock', regex = True)] # check through type 1 for only Grass or Rock
#df.loc[df['Name'].str.contains('pi(a-z)*',flags = re.I,regex=True)] # for displaying the names starting with pi 
# but caterpie is showing in the begining so for removing or for absolute words starting with pi we give ^ command
df.loc[df['Name'].str.contains('^pi(a-z)*',flags = re.I,regex=True)] # re.I for ignoring the capitals,* means zero or more
```

## Aggregrate Statistics(Group by)

```python
df.groupby(df['Type 1']).mean().sort_values('Speed',ascending=False) # grouping by certain column
# 3 statistic operations you can do by group by is SUM,MEAN,COUNT
```

```python
df.groupby(df['Type 1']).sum()
```

```python
df.groupby(df['Type 1']).count()
```

```python
# to get onlu count of each coulumn then 
df['count'] =1
df.groupby(df['Type 1']).count()['count']
```
